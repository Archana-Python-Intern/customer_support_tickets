```python
import re

# Task 2: Function to clean the ticket description
def clean_text(text):
    # 1. Convert to string and then to lowercase
    text = str(text).lower()
    # 2. Remove special characters and punctuation
    text = re.sub(r'[^a-z\s]', '', text)
    # 3. Remove extra whitespace
    text = text.strip()
    return text

# Apply the cleaning function to the 'Ticket Description' column
# This creates a new column called 'Cleaned_Description'
df['Cleaned_Description'] = df['Ticket Description'].apply(clean_text)

print("Task 2 Complete: Data Cleaned!")
# Show the first 5 rows to see the difference
df[['Ticket Description', 'Cleaned_Description']].head()
```

```python
import re

# Task 2: Function to clean the ticket description
def clean_text(text):
    # 1. Convert to string and then to lowercase
    text = str(text).lower()
    # 2. Remove special characters and punctuation
    text = re.sub(r'[^a-z\s]', '', text)
    # 3. Remove extra whitespace
    text = text.strip()
    return text

# Apply the cleaning function to the 'Ticket Description' column
# This creates a new column called 'Cleaned_Description'
df['Cleaned_Description'] = df['Ticket Description'].apply(clean_text)

print("Task 2 Complete: Data Cleaned!")
# Show the first 5 rows to see the difference
df[['Ticket Description', 'Cleaned_Description']].head()
```

```python
# Task 3: Define the classification function
def classify_ticket(text):
    text = str(text)

    # Logic for PAYMENT
    if any(word in text for word in ['pay', 'bill', 'charge', 'invoice', 'card', 'money']):
        return 'PAYMENT'
    # Logic for LOGIN
    elif any(word in text for word in ['login', 'password', 'account', 'access', 'sign']):
        return 'LOGIN'
    # Logic for DELIVERY
    elif any(word in text for word in ['delivery', 'shipping', 'track', 'order', 'package']):
        return 'DELIVERY'
    # Logic for REFUND
    elif any(word in text for word in ['refund', 'money back', 'return']):
        return 'REFUND'
    # Logic for BUG
    elif any(word in text for word in ['bug', 'error', 'crash', 'not working', 'issue']):
        return 'BUG'
    # Default category
    else:
        return 'GENERAL'

# Apply the function to the Cleaned_Description column
df['Category'] = df['Cleaned_Description'].apply(classify_ticket)

print("Task 3 Complete: Tickets Categorized!")
# Check the first 10 results
df[['Cleaned_Description', 'Category']].head(10)
```

```python
# Task 4: Priority Assignment Logic
def assign_priority(row):
    category = row['Category']

    # P0: Critical - Payment and Login issues
    if category == 'PAYMENT' or category == 'LOGIN':
        return 'P0'
    # P1: High - Technical bugs
    elif category == 'BUG':
        return 'P1'
    # P2: Medium - Shipping and returns
    elif category == 'DELIVERY' or category == 'REFUND':
        return 'P2'
    # P3: Low - General questions
    else:
        return 'P3'

# Apply priority to the dataframe
df['Priority'] = df.apply(assign_priority, axis=1)

print("Task 4 Complete: Priorities Assigned!")
df[['Category', 'Priority']].head(10)
```

```python
# Task 5: SLA Calculation
from datetime import timedelta

# Define SLA hours mapping
priority_sla = {'P0': 4, 'P1': 24, 'P2': 48, 'P3': 72}

# Map hours to the dataframe
df['SLA_Hours'] = df['Priority'].map(priority_sla)

# Convert 'Date of Purchase' to datetime objects
df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'])

# Calculate Due Date
df['Due_Date'] = df['Date of Purchase'] + pd.to_timedelta(df['SLA_Hours'], unit='h')

print("Task 5 Complete: SLA and Due Dates calculated!")
df[['Priority', 'SLA_Hours', 'Due_Date']].head(10)
```

```python
# Task 6: Report and Export
print("--- SUPPORT MANAGER REPORT ---")
print("\nTickets by Category:\n", df['Category'].value_counts())
print("\nTickets by Priority:\n", df['Priority'].value_counts())

# Export to CSV
df.to_csv('final_customer_support_triage.csv', index=False)
print("\nTask 6 Complete: Final CSV exported as 'final_customer_support_triage.csv'")
```
